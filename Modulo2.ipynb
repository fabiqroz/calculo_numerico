## O que são sinais e sistemas?? <a class="anchor" id="section_1_1"></a>

Basicamente, 
1. Sinais são **funções de uma ou mais variáveis independentes** que **contêm informações sobre o comportamento ou natureza
de algum fenômeno**.
2. **Sistemas respondem a sinais** em particular, **produzindo outros sinais ou algum comportamento** desejado.

## Sinais contínuos e discretos no tempo <a class="anchor" id="section_2_1"></a>

No caso dos sinais de tempo contínuo, a **variável independente é contínua** e, portanto, esses sinais são
definidos em um conjunto contínuo de valores da variável independente. Em contrapartida, os sinais de tempo
discreto são definidos somente em instantes discretos, ou seja, a **variável independente assume apenas um conjunto discreto de valores**.

Exemplos:

1. Um sinal de fala em função do tempo e a pressão atmosférica em função da altitude são exemplos de sinais de tempo contínuo. 
2. O índice semanal Dow-Jones da Bolsa de Valores de Nova York é um exemplo de sinal de tempo discreto

<div> <img src="figures/sinal_continuo.png" alt="Drawing" style="width: 400px;"/></div> 

<div> <img src="figures/sinal_discreto.png" alt="Drawing" style="width: 600px;"/></div> 

Um sinal de tempo discreto $x[n]$ pode representar um fenômeno para o qual a variável independente é inerentemente discreta. Por outro lado, **uma classe muito importante de sinais de tempo discreto decorre da amostragem de sinais de tempo contínuo**. Nesse caso, o sinal de tempo discreto $x[n]$ representa amostras sucessivas de um fenômeno para o qual a variável independente é contínua.

Um sinal **discreto no tempo**  pode ser *inerentemente discreto* ou corresponder a uma versão *amostrada* de um sinal de tempo contínuo. Um exemplo do primeiro seria um sinal correspondente ao mercado de ações Dow Jones Industrial Average
índice (que é definido apenas em intervalos diários), enquanto um exemplo do último seria a versão amostrada de um (tempo contínuo) sinal de fala.

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
import numpy as np
```

```python
# Taxa de amostragem do sinal que iremos gerar
n = 50000

# Amostrando o sinal no tempo
x = np.linspace(-10, 10, n, endpoint=False)

#Função que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
def sinal(x):
    return np.power(x,3) - np.power(x,2)/2 - 2*x
    #return np.power(x,2)


# Vizualização
plt.plot(x, sinal(x))
plt.show()
```

## Sinais analógicos e digitais <a class="anchor" id="section_2_2"></a>

O conceito de tempo contínuo é geralmente confundido com o conceito de analógico. **Contudo, os dois conceitos são diferentes**. O mesmo é válido para os conceitos de tempo discreto e digital. 

* Os termos contínuo no tempo e discreto no tempo qualificam a natureza do sinal ao longo do eixo de tempo (eixo horizontal). 
* Os termos analógico e digital, por outro lado, qualificam a natureza da amplitude do sinal (eixo vertical).


Como já previamente mencionado, o processo de conversão de sinais contínuos em discretos se chama **Amostragem**.  Já o processo de conversão de sinais analógicos em digitais se chama **Quantização**

## Sinais Periódicos e não-periódicos <a class="anchor" id="section_2_3"></a>

Um sinal $x(t)$ é dito periódico se para alguma constante positiva $T$ $$x(t) = x(t+T) \;\; \text{para todo t} $$

Em outras palavras, um sinal periódico tem a propriedade de não se modificar pelo deslocamento no tempo de $T$.

* O menor valor de $T$ que satisfaça a equação é o **período fundamental** do sinal $x(t)$

<div> <img src="figures/periodico.png"/></div> 

```python
def periodic_function(func, period):
    return lambda x: func(((x ) % period )  )

# Taxa de amostragem do sinal que iremos gerar
n = 100
T = 10

# # Amostrando o sinal no tempo
a = np.linspace(0, 30, n, endpoint=False)

sinal = periodic_function(lambda x: x, T)

plt.plot(a,sinal(a))
plt.show()
```

Um sinal que não começar antes de $t = 0$ é um **sinal causal**. Em outras palavras, x(t) é um sinal causal se:
$$x(t) = 0 \;\;\; t < 0$$

Sinais Causais:

<div> <img src="figures/causais.png" alt="Drawing" style="width: 600px;"/></div> 

Sinais não-causais:

<div> <img src="figures/nao_causais.png" alt="Drawing" style="width: 600px;"/></div> 

```python
# Gera sinal causal com frequência de F Hz
def sinal_senoidal(periodo, frequencia, tempo):
    sinal = np.sin(periodo*frequencia*tempo)
    return sinal

# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
x = np.linspace(0, 5, n, endpoint=False)

# Definindo o período e frequencia do sinal
p = 4*np.pi
f = 1

# Vizualização
plt.plot(x, sinal_senoidal(p,f,x))
plt.show()
```

## Sinais determinísticos e estocásticos <a class="anchor" id="section_2_5"></a>

Um sinal cuja descrição física é completamente conhecida, seja na forma matemática ou na forma gráfica é um sinal determinístico. Um sinal cujos valores não podem ser preditos precisamente, mas são conhecidos apenas em termos de uma descrição probabilística, tal como o valor médio ou valor médio quadrático, são sinais aleatórios. Neste livro, trabalharemos exclusivamente com sinais determinísticos. Os sinais aleatórios estão além do escopo dos nosssos estudos.

```python
# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
x = np.linspace(1, 100, n, endpoint=False)

#Função que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
def sinal_aleatorio(x):
    return np.sin(x) + np.random.normal(scale=0.8, size=len(x))

# Vizualização
plt.plot(x, sinal_aleatorio(x))
plt.show()
```

## Energia e potência de um sinal <a class="anchor" id="section_2_4"></a>

### Tamanho de um Sinal <a class="anchor" id="section_2_4_1"></a>

O tamanho de qualquer entidade é um número que indica a largura ou o comprimento da entidade. Genericamente falando, a amplitude do sinal varia com o tempo. **Como um sinal que existe em um certo intervalo de tempo com amplitude variante pode ser medido por um número que irá indicar o tamanho ou a força do sinal?** Tal
medida deve considerar não apenas a amplitude do sinal, mas também sua duração. 

Por exemplo, se quisermos utilizar um único número V como medida do tamanho de um ser humano, devemos considerar não somente seu peso, mas também sua altura. Se fizermos uma consideração que a forma da pessoa é um cilindro cuja variável é o raio r (o qual varia com a altura h), então uma possível medida do tamanho de uma pessoa de altura H é o volume V da pessoa, dado por $$V = \int_o^H r^2(h)dh$$

### Energia de um Sinal <a class="anchor" id="section_2_4_2"></a>

Argumentando desta forma, podemos considerar a área abaixo do sinal $x(t)$ como uma possível medida de seu tamanho, pois a área irá considerar não somente a amplitude, mas também sua duração.$$Tamanho = \int_{-\infty}^{\infty} x(t)dt$$ 


```python
# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
x = np.linspace(-10, 10, n, endpoint=False)

#Função que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
sinal = np.sin(x)

#vizualização
plt.rcParams["figure.figsize"] = [7.50, 3.50]

plt.fill_between(x, sinal, step="pre", alpha=0.4)
plt.plot(x, sinal, drawstyle="steps")
plt.xlabel('tempo (t)')
plt.ylabel('Amplitude  x(t)')

plt.show()
```

Entretanto, esta medida ainda é defeituosa, pois mesmo para um sinal grande $x(t)$, suas áreas positivas e negativas podem se cancelar, indicando um sinal de tamanho pequeno:

Esta dificuldade pode ser corrigida pela definição do tamanho do sinal como a área debaixo de $x^2(t)$ a qual é sempre positiva. Podemos chamar essa medida de **energia do sinal**: $$E = \int_{-\infty}^{\infty} x^2(t)dt\$$

```python

#Função que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
sinal = np.power(np.sin(x),2)

#vizualização
plt.rcParams["figure.figsize"] = [7.50, 3.50]

plt.fill_between(x, sinal, step="pre", alpha=0.4)
plt.plot(x, sinal, drawstyle="steps")
plt.xlabel('tempo (t)')
plt.ylabel('Amplitude  x(t)')

plt.show()
```

Essa definição pode ser generalizada para um sinal complexo $x(t)$, sendo dada por $$E = \int_{-\infty}^{\infty} |x(t)|^2 dt$$

**A energia do sinal deve ser finita** para que seja uma medida significativa do tamanho do sinal. 
Uma condição necessária para que a energia seja finita é que a amplitude do sinal → 0 quando |t|→ ∞. Caso contrário a integral da equação acima não irá convergir.

### Potencia de um sinal  <a class="anchor" id="section_2_4_3"></a>

Quando a amplitude do sinal x(t) não → 0 quando |t|→∞, a energia do sinal é infinita. 

<div> <img src="figures/s2.png"/></div> 

Uma medida mais significativa do tamanho do sinal neste caso é a energia média, se ela existir. Esta medida é chamada de **potência do sinal**. Para um sinal $x(t)$, definimos sua potência $P$ por $$P = \lim_{T \rightarrow \infty} \int_{-T/2}^{T/2} |x(t)|^2 dt$$

### Observando a energia e potência de um sinal que representa um fenômeno físico:

Em muitas aplicações, os sinais que consideramos estão diretamente relacionados a **quantidades físicas** e a partir deles pode-se extrair a **potência ou energia de um sistema físico**.  

Por exemplo, se $v(t)$ e $i(t)$ são, respectivamente, a tensão e a corrente através de um resistor com resistência $R$, então a potência instantânea é 

$$p(t) = v(t)i(t) = \frac{1}{R}v^2(t)$$

A energia total dissipada no intervalo de tempo $t_1\le t\le t_2$ é 
$$\int_{t_1}^{t_2}p(t)dt = \int_{t_1}^{t_2}\frac{1}{R}v^2(t)dt $$

e a potência média durante esse intervalo de tempo é 

$$\frac{1}{t_2 - t_1}\int_{t_1}^{t_2}p(t)dt = \frac{1}{t_2 - t_1}\int_{t_1}^{t_2}\frac{1}{R}v^2(t)dt$$
